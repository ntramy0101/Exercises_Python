```python
"""============================================================================
1 Viết function truyền vào số nguyên dương n. Trả về list bình phương của
các số từ 1->n.
VD: 
input: 4
output: [1, 4, 9, 16]
"""
def tinh_binh_phuong(n):
    """
    Hàm này trả về danh sách các số bình phương từ 1 đến n.
    Args:
    - n: Số nguyên dương đầu vào.
    Returns:
    - Danh sách bình phương từ 1 đến n.
    """
    if n < 1:
        return []  # Trả về danh sách rỗng nếu n không hợp lệ
    b_squared = [i**2 for i in range(1, n + 1)]
    return b_squared
# Nhập số từ bàn phím
n = int(input("Nhập một số nguyên dương: "))
ket_qua = tinh_binh_phuong(n)
print(f"Danh sách bình phương từ 1 đến {n}: {ket_qua}")
```

```python
"""============================================================================
2. Viết function truyền vào 2 tham số m:số dân hiện tại và n:số năm
trả về dân số của 1 nước sau n năm, tì lệ tăng dân số là 1.8% hàng năm
"""
def tinh_dan_so_du_doan(dan_so_hien_tai, so_nam, ty_le_tang_dan_so=1.8):
    """
    Hàm này tính dự đoán dân số của một quốc gia sau một số năm với tỷ lệ tăng dân số hàng năm.
    Args:
    - dan_so_hien_tai: Số dân hiện tại.
    - so_nam: Số năm dự đoán.
    - ty_le_tang_dan_so: Tỷ lệ tăng dân số hàng năm (tính bằng phần trăm), giá trị mặc định là 1.8%.
    Returns:
    - Dự đoán dân số sau số năm cho trước.
    """
    if dan_so_hien_tai < 0 or so_nam < 0 or ty_le_tang_dan_so < 0:
        return "Không hợp lệ"
    # Tính dân số dự đoán sau số năm
    dan_so_du_doan = dan_so_hien_tai * (1 + ty_le_tang_dan_so / 100) ** so_nam
    return int(dan_so_du_doan)
while True:
    dan_so_hien_tai = int(input("Nhập số dân hiện tại: "))
    so_nam = int(input("Nhập số năm dự đoán: "))
    ket_qua = tinh_dan_so_du_doan(dan_so_hien_tai, so_nam)
    print(f"Dân số sau {so_nam} năm dự đoán là: {ket_qua}")
    tiep_tuc = input("Tiếp tục dự đoán? (Nhấn Enter để tiếp tục, hoặc nhập 'q' để thoát): ")
    if tiep_tuc.lower() == 'q':
        break
```

```python
"""============================================================================
3. Viết function cho người dùng nhập chỉ số điện mới và cũ từ bàn phím.
    In ra màn hình chỉ số cũ, 
    chỉ số mới, 
    tiền trả định mức, 
    tiền trả vượt định mức, 
    tổng tiền phải trả. 

Định mức sử dụng điện cho mỗi hộ là: 50 KW với giá 2300đ/KW
Nếu phần vượt định mức <= 50KW thì tính giá 4800đ/KW
Nếu 50KW < phần vượt định mức < 100KW thì tính giá 7000đ/KW 
Nếu phần vượt định mức >= 100KW thì tính giá 9000đ/KW 
"""

def tinh_tien_dien(so_dien_cu, so_dien_moi):
    # Định mức sử dụng điện cho mỗi hộ
    dinh_muc = 50  # KW
    gia_dinh_muc = 2300  # đồng/KW
    gia_vuot_dinh_muc_1 = 4800  # đồng/KW
    gia_vuot_dinh_muc_2 = 7000  # đồng/KW
    gia_vuot_dinh_muc_3 = 9000  # đồng/KW
    # Tính số điện sử dụng và tiền điện
    so_dien_su_dung = so_dien_moi - so_dien_cu
    tien_dien = 0
    if so_dien_su_dung <= dinh_muc:
        tien_dien = so_dien_su_dung * gia_dinh_muc
    elif so_dien_su_dung <= 2 * dinh_muc:
        tien_dien = dinh_muc * gia_dinh_muc + \
            (so_dien_su_dung - dinh_muc) * gia_vuot_dinh_muc_1
    elif so_dien_su_dung <= 3 * dinh_muc:
        tien_dien = dinh_muc * gia_dinh_muc + dinh_muc * gia_vuot_dinh_muc_1 + \
            (so_dien_su_dung - 2 * dinh_muc) * gia_vuot_dinh_muc_2
    else:
        tien_dien = dinh_muc * gia_dinh_muc + dinh_muc * gia_vuot_dinh_muc_1 + dinh_muc * \
             gia_vuot_dinh_muc_2 + \
            (so_dien_su_dung - 3 * dinh_muc) * gia_vuot_dinh_muc_3
    return so_dien_cu, so_dien_moi, so_dien_su_dung, tien_dien
# Nhập số điện cũ và số điện mới từ người dùng
so_dien_cu = int(input("Nhập số điện cũ: "))
so_dien_moi = int(input("Nhập số điện mới: "))
# Tính tiền điện và in ra kết quả
so_dien_cu, so_dien_moi, so_dien_su_dung, tien_dien = tinh_tien_dien(
    so_dien_cu, so_dien_moi)
print(f"Số điện cũ: {so_dien_cu} KW")
print(f"Số điện mới: {so_dien_moi} KW")
print(f"Số điện sử dụng: {so_dien_su_dung} KW")
print(f"Số tiền phải trả: {tien_dien} đồng")
```

```python
"""============================================================================
4. Viết function trả về kết quả xếp loại học lực của học sinh. 
Giả sử học sinh chỉ có bốn cột điểm Toán (T), Lý (L), Hóa (H) và Anh văn (AV). 
Điểm 4 môn được nhập vào từ bàn phím và theo thang điểm 100. 

Qui tắc xếp loại như sau:
    + Nếu trung bình 4 môn lớn hơn hoặc bằng 80 và không có môn nào nhỏ hơn 65, 
    thì xếp loại Giỏi.
    + Nếu học sinh không đủ điều kiện đạt loại giỏi, mà có trung bình 4 môn 
    lớn hơn hoặc bằng 65 và không có môn nào nhỏ hơn 50, thì xếp loại Khá.
    + Nếu học sinh không đủ điều kiện đạt loại Khá, mà có trung bình 4 môn 
    lớn hơn hoặc bằng 50 và không có môn nào nhỏ hơn 20, thì xếp loại 
    Trung bình.
    + Nếu học sinh không đạt 3 loại trên thì xếp loại Yếu.
"""
def xep_loai_hoc_luc(toan, ly, hoa, anh_van):
    # Tính điểm trung bình 4 môn
    diem_trung_binh = (toan + ly + hoa + anh_van) / 4
    # Xếp loại học lực
    if diem_trung_binh >= 80 and toan >= 65 and ly >= 65 and hoa >= 65 and anh_van >= 65:
        return "Giỏi"
    elif diem_trung_binh >= 65 and toan >= 50 and ly >= 50 and hoa >= 50 and anh_van >= 50:
        return "Khá"
    elif diem_trung_binh >= 50 and toan >= 20 and ly >= 20 and hoa >= 20 and anh_van >= 20:
        return "Trung bình"
    else:
        return "Yếu"
# Nhập điểm số từ bàn phím
toan = float(input("Nhập điểm Toán: "))
ly = float(input("Nhập điểm Lý: "))
hoa = float(input("Nhập điểm Hóa: "))
anh_van = float(input("Nhập điểm Anh Văn: "))
# Xếp loại học lực và in ra kết quả
xep_loai = xep_loai_hoc_luc(toan, ly, hoa, anh_van)
#In ra điểm đã nhập 
print(f"Điểm Toán: {toan}")
print(f"Điểm Lý: {ly}")
print(f"Điểm Hóa: {hoa}")
print(f"Điểm Anh văn: {anh_van}")
#In ra kết quả học lực
print(f"Xếp loại học lực: {xep_loai}")
```

```python
"""============================================================================
5. Viêt function truyền vào 2 số A < B < 2000, trả về danh sách chứa tất cả
các số chia hết cho 7 nhưng không chia hết cho 5, nằm trong đoạn từ A 
đến B(tính cả A và B)
"""
def danh_sach_so_chia_het_cho_7_nhung_khong_chia_het_cho_5(A, B):
    danh_sach = []
    for i in range(A, B + 1):
        if i % 7 == 0 and i % 5 != 0:
            danh_sach.append(i)
    return danh_sach
# Nhập giá trị A và B từ bàn phím
A = int(input("Nhập giá trị A (A < B < 2000): "))
B = int(input("Nhập giá trị B (A < B < 2000): "))

# Kiểm tra điều kiện A < B < 2000
if A < B < 2000:
    ket_qua = danh_sach_so_chia_het_cho_7_nhung_khong_chia_het_cho_5(A, B)
    print(
        f"Danh sách các số chia hết cho 7 nhưng không chia hết cho 5 trong khoảng từ {A} đến {B}:")
    print(ket_qua)
else:
    print("Giá trị A hoặc B không thỏa mãn điều kiện A < B < 2000.")
```

```python
"""============================================================================
6. Viết function in ra màn hình xem chuỗi kí tự nhập vào từ người dùng có hợp
lệ hay không. Chuỗi hợp lệ là chuỗi có chiều dài ít nhất 8 kí tự, chứa ít nhất
một chữ in hoa, chứa ít nhất một kí tự dạng số.
"""
def kiem_tra_chuoi_hop_le(chuoi):
    # Kiểm tra chiều dài của chuỗi
    if len(chuoi) < 8:
        return False
    # Kiểm tra xem chuỗi có ít nhất một chữ in hoa
    if not any(char.isupper() for char in chuoi):
        return False
    # Kiểm tra xem chuỗi có ít nhất một kí tự dạng số
    if not any(char.isdigit() for char in chuoi):
        return False
    # Nếu tất cả các điều kiện đều đúng, trả về True
    return True
# Nhập chuỗi từ người dùng
chuoi_nhap = input("Nhập một chuỗi: ")

print(f"Chuỗi nhập {chuoi_nhap}")
while:
# Kiểm tra chuỗi nhập từ người dùng
if kiem_tra_chuoi_hop_le(chuoi_nhap):
    print("Chuỗi hợp lệ.")
else:
    print("Chuỗi không hợp lệ.")
    print("Điều kiện chuỗi hợp lệ: Chuỗi hợp lệ là chuỗi có chiều dài ít nhất 8 kí tự, chứa ít nhất một chữ in hoa, chứa ít nhất một kí tự dạng số.")
```

```python
"""============================================================================
10. Viết function tính tần suất xuất hiện các từ(cách nhau bởi dấu cách " ") 
từ input. In ra màn hình kết quả sau khi đã sắp xếp theo bảng chữ cái.
VD Input: 
New to Python or choosing between Python 2 and Python 3? Read Python 2 or 
Python 3.
- Output: 2:2 , 3.:1 , 3?:1 , New:1 , Python:5 , Read:1 , and:1 , between:1 ,
 choosing:1 , or:2 , to:1
"""
def tinh_tan_suat(input_str):
    # Chuyển chuỗi đầu vào thành danh sách các từ
    words = input_str.split()
    # Tạo một dictionary để lưu tần suất xuất hiện của các từ
    word_freq = {}
    for word in words:
        # Bỏ dấu câu và chuyển thành chữ thường để đếm tần suất
        cleaned_word = word.strip('.,?').lower()
        if cleaned_word in word_freq:
            word_freq[cleaned_word] += 1
        else:
            word_freq[cleaned_word] = 1
    # Sắp xếp dictionary theo thứ tự bảng chữ cái
    sorted_word_freq = dict(sorted(word_freq.items()))
    # In kết quả
    for word, freq in sorted_word_freq.items():
        print(f"{word}:{freq}", end=" , ")
input_str = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."
tinh_tan_suat(input_str)
```

```python
"""============================================================================
11.Viết module triangle trong đó có các hàm:
- kiểm tra xem tam giác là tam giác cân, tam giác đều, tam giác vuông 
hay thường?
- tính chu vi tam giác
- tính diện tính tam giác.
input các hàm trên là chiều dài 3 cạnh tam giác.
"""
def classify_triangle(a, b, c):
    # Sắp xếp các cạnh để dễ dàng kiểm tra tam giác vuông
    # sorted: sắp xếp
    a, b, c = sorted([a, b, c])
    
    if a + b <= c:
        return "Không phải tam giác"
    elif a == b == c:
        return "Tam giác đều"
    elif a == b or b == c:
        return "Tam giác cân"
    elif a**2 + b**2 == c**2:
        return "Tam giác vuông"
    else:
        return "Tam giác thường"

#chu vi
def perimeter(a, b, c):
    return a + b + c

#diện tích
def area(a, b, c):
    s = (a + b + c) / 2
    return (s * (s - a) * (s - b) * (s - c))**0.5

# Test the functions
a = float(input("Nhập độ dài cạnh a: "))
b = float(input("Nhập độ dài cạnh b: "))
c = float(input("Nhập độ dài cạnh c: "))

triangle_type = classify_triangle(a, b, c)
print("Loại tam giác:", triangle_type)

if triangle_type != "Không phải tam giác":
    print("Chu vi tam giác là:", perimeter(a, b, c))
    print("Diện tích tam giác là:", area(a, b, c))
```

```python
"""============================================================================
12. Xây dựng function trả về ma trận đơn vị kích thước N*N
VD Input: N = 3
Ouput = [[1,0,0],[0,1,0],[0,0,1]]
"""
def tao_ma_tran_don_vi(N):
    # Khởi tạo một danh sách trống để lưu trữ ma trận đơn vị
    ma_tran_don_vi = []

    # Duyệt qua từng hàng của ma trận
    for i in range(N):
        # Khởi tạo một hàng mới (list) với độ dài N, ban đầu chứa toàn số 0
        hang_moi = [0] * N

        # Đặt giá trị của đường chéo chính (các phần tử có cùng chỉ số hàng và cột) thành 1
        hang_moi[i] = 1

        # Thêm hàng này vào ma trận đơn vị
        ma_tran_don_vi.append(hang_moi)

    # Trả về ma trận đơn vị đã tạo
    return ma_tran_don_vi

# Nhập kích thước ma trận từ người dùng
N = int(input("Nhập kích thước ma trận (N*N): "))

# Gọi hàm để tạo ma trận đơn vị
ma_tran_don_vi = tao_ma_tran_don_vi(N)

# In ma trận đơn vị
print("Ma trận đơn vị kích thước", N, "*", N, "là:")
for hang in ma_tran_don_vi:
    print(hang)
```

```python
"""============================================================================
13. Xây dựng module mymath.py gồm các function sau:
- xuất ra giá trị xuất hiện nhiều nhất trong 1 list số, trả về nhiều giá trị 
nếu có.(gọi ý dùng mode() trong statistics)
VD: Input:[1,2,5,7,3,7,7,5,8,5]
Output: 5, 7

- tính giai thừa 1 số nguyên.
- kiểm tra 1 số có phải là số nguyên tố hay không? 
(số nguyên tố là số chỉ chia hết cho 1 và chính nó.)
- kiểm tra 1 số có phải số chính phương hay không? 
(số chính phương là số có căn bậc 2 là số tự nhiên.)
Tạo 1 file main.py để gọi hàm từ module mymath
"""

""" Module mymath.py:============================================================="""

import statistics
import math

# Hàm để tìm giá trị xuất hiện nhiều nhất trong một danh sách số
def tim_gia_tri_xuat_hien_nhieu_nhat(danh_sach):
    try:
        # Sử dụng hàm mode từ thư viện statistics để tìm giá trị xuất hiện nhiều nhất
        mode_values = statistics.mode(danh_sach)
        return mode_values
    except statistics.StatisticsError:
        # Trong trường hợp không có giá trị xuất hiện nhiều nhất
        return None

# Hàm tính giai thừa của một số nguyên
def tinh_giai_thua(n):
    if n == 0:
        return 1
    else:
        return n * tinh_giai_thua(n - 1)

# Hàm kiểm tra xem một số có phải là số nguyên tố hay không
def la_so_nguyen_to(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Hàm kiểm tra xem một số có phải là số chính phương hay không
def la_so_chinh_phuong(n):
    if n < 0:
        return False
    sqrt_n = int(math.sqrt(n))
    return sqrt_n * sqrt_n == n


""" Tệp main.py:========= """

import mymath

# Danh sách số để kiểm tra giá trị xuất hiện nhiều nhất
danh_sach = [1, 2, 5, 7, 3, 7, 7, 5, 8, 5]
# Gọi hàm từ module mymath để tìm giá trị xuất hiện nhiều nhất
kq = mymath.tim_gia_tri_xuat_hien_nhieu_nhat(danh_sach)
if kq:
    print("Giá trị xuất hiện nhiều nhất trong danh sách:", kq)
else:
    print("Không có giá trị nào xuất hiện nhiều nhất trong danh sách.")

# Số nguyên để tính giai thừa
n = 5
# Gọi hàm từ module mymath để tính giai thừa
kq = mymath.tinh_giai_thua(n)
print(f"{n}! = {kq}")

# Số để kiểm tra xem có phải số nguyên tố
so = 17
# Gọi hàm từ module mymath để kiểm tra số nguyên tố
if mymath.la_so_nguyen_to(so):
    print(f"{so} là số nguyên tố.")
else:
    print(f"{so} không phải là số nguyên tố.")

# Số để kiểm tra xem có phải số chính phương
so = 16
# Gọi hàm từ module mymath để kiểm tra số chính phương
if mymath.la_so_chinh_phuong(so):
    print(f"{so} là số chính phương.")
else:
    print(f"{so} không phải là số chính phương.")
```

```python
"""============================================================================
15.
Viết hàm nhận vào file transaction.txt với dòng đầu tiên là id của tài khoản, 
dòng thứ 2 là số tài khoản, dòng thứ 3 là số dư tài khoản, các dòng tiếp theo 
hiển thị nội dung giao dịch 
    + Với D,100 là nộp vào ngân hàng 100. 
    + Với W,100 là rút ra 100 từ tài khoản. 
Hàm trả về số dư cuối cùng của tài khoản
vd input:
Id1
Tk01
400
D,100
W,200
-> output: 300
"""
def process_transactions(filename):
    # Khởi tạo biến số dư ban đầu
    balance = 0

    try:
        # Mở file transaction.txt để đọc
        with open(filename, 'r') as file:
            # Đọc id tài khoản
            account_id = file.readline().strip()

            # Đọc số tài khoản
            account_number = file.readline().strip()

            # Đọc số dư tài khoản
            balance = float(file.readline().strip())

            # Đọc các giao dịch từ file
            for line in file:
                # Tách thông tin giao dịch thành loại và số tiền
                transaction_type, amount = line.strip().split(',')

                # Chuyển số tiền thành kiểu số
                amount = float(amount)

                # Xử lý giao dịch nộp tiền (D) hoặc rút tiền (W)
                if transaction_type == 'D':
                    balance += amount
                elif transaction_type == 'W':
                    balance -= amount

    except FileNotFoundError:
        print("File not found!")

    return balance

# Đường dẫn tới file transaction.txt
file_path = "transaction.txt"

# Gọi hàm và in ra số dư cuối cùng của tài khoản
final_balance = process_transactions(file_path)
print("Số dư cuối cùng của tài khoản:", final_balance)
```

```python
"""============================================================================
17.
Một Robot di chuyển trong mặt phẳng bắt đầu từ điểm đầu tiên. 
Robot có thể di chuyển theo hướng UP, DOWN, LEFT và RIGHT với những bước 
nhất định. Dấu di chuyển của robot được lấy từ file txt đầu vào, với 2 dòng 
đầu tiên thể hiện vị trí điểm đầu tiên, các dòng tiếp theo thể hiện quá trình 
di chuyển trong đó các con số phía sau hướng di chuyển chính là số bước đi 
VD: Input: 
X0 4 
Y0 3 
UP 5 
DOWN 3 
LEFT 3 
RIGHT 3 

Hãy viết hàm trả về khoảng cách từ vị trí hiện tại đến vị trí đầu tiên, 
sau khi robot đã di chuyển một quãng đường với input là tên file txt.
"""

def calculate_distance(filename):
    # Khởi tạo biến vị trí hiện tại của robot
    current_position = [0, 0]

    try:
        # Mở file để đọc
        with open(filename, 'r') as file:
            for line in file:
                # Tách dòng thành các từ
                words = line.strip().split()
                # Nếu dòng chứa tọa độ ban đầu của robot
                if words[0] == "X0":
                    current_position[0] = int(words[1])
                elif words[0] == "Y0":
                    current_position[1] = int(words[1])
                # Nếu dòng chứa hướng di chuyển và số bước
                else:
                    direction = words[0]
                    steps = int(words[1])
                    # Cập nhật vị trí hiện tại dựa trên hướng di chuyển và số bước
                    if direction == "UP":
                        current_position[1] += steps
                    elif direction == "DOWN":
                        current_position[1] -= steps
                    elif direction == "LEFT":
                        current_position[0] -= steps
                    elif direction == "RIGHT":
                        current_position[0] += steps

    except FileNotFoundError:
        print("File not found!")

    # Tính khoảng cách từ vị trí hiện tại đến vị trí ban đầu
    distance = ((current_position[0] - 0) ** 2 + (current_position[1] - 0) ** 2) ** 0.5
    return distance

# Đường dẫn tới file văn bản
file_path = "robot_movement.txt"

# Gọi hàm và in ra khoảng cách
result = calculate_distance(file_path)
print("Khoảng cách từ vị trí hiện tại đến vị trí ban đầu:", result)

"""
Trong ví dụ trên, chúng ta đọc từng dòng trong file văn bản, tách dòng thành các từ, 
và kiểm tra từng dòng để xác định tọa độ ban đầu của robot hoặc hướng di chuyển và số bước. 
Sau khi xử lý hết file, chúng ta tính toán khoảng cách từ vị trí hiện tại đến vị trí ban đầu 
bằng công thức Euclid và trả về kết quả.
"""
```

```python
"""===========================================================================
18. Tìm số nguyên dương N bé nhất sao cho N chia hết cho tất cả các số trong 
đoạn [1, 20]
"""

def tim_boi_so_chung_nho_nhat():
    # Sử dụng thư viện math để tính GCD (ước chung lớn nhất)
    from math import gcd

    # Khởi tạo N ban đầu bằng 1
    N = 1

    # Tìm bội số chung nhỏ nhất của các số từ 1 đến 20
    for i in range(1, 21):
        N = (N * i) // gcd(N, i)

    return N

# Gọi hàm để tìm số N
N = tim_boi_so_chung_nho_nhat()
print("Số nguyên dương nhỏ nhất chia hết cho tất cả các số từ 1 đến 20 là:", N)
```
