```python
'''
Ex.8) load các data sau và thực hiện theo yêu cầu:
week1 = pd.read_csv(data_path + "Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv(data_path + "Restaurant - Week 2 Sales.csv")
customers = pd.read_csv(data_path + "Restaurant - Customers.csv")
foods = pd.read_csv(data_path + "Restaurant - Foods.csv")

1) Join all data for analysis:
    - Concat data week1 and week2
    - Merge with data foods & customers
2) Statistic & visualize features of the data

3) Khảo sát 
  - Tỉ lệ nam nữ
  - Food theo giới tính
  - Giá theo giới tính
  - Company:
    + Tỉ lệ nam nữ đi mua hàng ở mỗi cty
    + Loại thức ăn mà các thành viên ở các cty hay mua
    + Có mua lặp lại không
    + Thống kê giá sản phẩm được mua bởi thành viên của các công ty
'''

import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ các tệp
data_path = "đường_dẫn_đến_thư_mục_chứa_tệp_csv/"
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")

# 1. Join data
# - Concat data week1 and week2
combined_weeks = pd.concat([week1, week2])
# - Merge with data foods & customers
combined_data = combined_weeks.merge(customers, on='Customer ID').merge(foods, on='Food ID')

# 2. Statistic & visualize features of the data

# 3. Khảo sát
# - Tỉ lệ nam nữ
gender_ratio = customers['Gender'].value_counts()
gender_ratio.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title("Gender Distribution")
plt.show()

# - Food theo giới tính
food_by_gender = combined_data.groupby(['Gender', 'Food Item']).size().unstack().fillna(0)
food_by_gender.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Food Ordered by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# - Giá theo giới tính
price_by_gender = combined_data.groupby('Gender')['Price Each'].mean()
price_by_gender.plot(kind='bar', color=['skyblue', 'lightcoral'], figsize=(10, 6))
plt.title("Average Price by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Price")
plt.xticks(rotation=0)
plt.show()

# - Company
#   + Tỉ lệ nam nữ đi mua hàng ở mỗi công ty
company_gender_ratio = customers.groupby(['Company', 'Gender']).size().unstack().fillna(0)
company_gender_ratio.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Gender Ratio by Company")
plt.xlabel("Company")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

#   + Loại thức ăn mà các thành viên ở các công ty hay mua
company_food = combined_data.groupby(['Company', 'Food Item']).size().unstack().fillna(0)
company_food.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Food Ordered by Company")
plt.xlabel("Company")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

#   + Có mua lặp lại không
company_repeat = customers.groupby('Company')['Customer ID'].nunique()
company_repeat.plot(kind='bar', color='lightcoral', figsize=(10, 6))
plt.title("Unique Customers by Company")
plt.xlabel("Company")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

#   + Thống kê giá sản phẩm được mua bởi thành viên của các công ty
company_price = combined_data.groupby('Company')['Price Each'].mean()
company_price.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title("Average Price by Company")
plt.xlabel("Company")
plt.ylabel("Average Price")
plt.xticks(rotation=0)
plt.show()
```

```python
import numpy as np
import mplcyberpunk
#tạo giá trị cho mảng x
x=np.linspace(0,10,40)
#tạo giá trị cho mảng y1, y2
y1=np.sin(x)
y2=np.cos(x)

#sử dụng style 'cyberpunk', tạo figure có kích thước (8,5)
plt.style.use ('cyberpunk')
plt.figure (figsize= (8,5))

# tạo biểu đồ scatter biểu diên sin(x) với các điểm phát sáng
# các điểm là các chấm tròn có bán kính 2 pixel (s=20)

plt.scatter (x, y1, marker= 'o', s=20)
mplcyberpunk.make_scatter_glow()

#tạo biểu đồ scatter biểu diễn cos(x) với các điểm phát sáng, màu lime
#trên cùng 1 figure

plt.scatter (x, y2, marker='o', c='lime', s=20)
mplcyberpunk.make_scatter_glow()
#hiển thị tên cho trục x, trục y là X - Axis, Y-axis
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
#hiển thị biểu đò
plt.show()

```
